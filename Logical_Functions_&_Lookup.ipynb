1. Write a formula to check if the marks in cell are greater than or equal to 40. If yes, return , otherwise B2 "Pass" "Fail".

2. A company gives a bonus if sales are above 40,000 and attendance is greater than or equal to 90%. Write a formula to calculate whether the employee will get a bonus or not.

3. Write a nested IF formula:If marks in are ≥ 90 → , ≥ 75 → , ≥ 50 → , otherwise .B2 "A" "B" "C" "Fail".

4. Find the Department of Employee ID = E103. Also explain which LOOKUP function is used and why?
  - Use VLOOKUP because EmpID is in the leftmost column and Department is to the right of it in the same table.
  formula (assuming EmpID-Department table is in A2:C5 and EmpID in F2):
  =VLOOKUP("E103",A2:C5,2,FALSE)
  VLOOKUP searches for "E103" in the first column of the range and returns the matching value from the 2nd column (Department).

  This returns the rate for the largest Sales Amount less than or equal to 45000.

7. Find the salary of EmpID = E105 using INDEX-MATCH Function.
